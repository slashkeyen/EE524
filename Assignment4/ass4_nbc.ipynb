```
import numpy as np
def fit(x,y):
  n_sample,n_feature=x.shape
  labels=np.unique(y)
  n_labels=len(labels)
  mean_x=np.zeros((n_labels,n_feature),dtype=np.float64)
  var_x=np.zeros((n_labels,n_feature),dtype=np.float64)
  prior=np.zeros(n_labels,dtype=np.float64)
  for k in labels:
    x_l=x[k==y]
    mean_x[k,:]=x_l.mean(axis=0)
    var_x[k,:]=x_l.var(axis=0)
    prior[k]=x_l.shape[0]/float(n_sample)
  return [mean_x,var_x,prior]


import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
iris=datasets.load_iris()
x=iris.data
y=iris.target
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0)

data_fit=fit(xtrain,ytrain)


def gau_pdf(i,x):
 m=data_fit[0][i]
 v=data_fit[1][i]
 exponent=np.exp(-(x-m)**2/(2*v))
 return (1/(2*np.pi*v))*exponent

def likelihood(k,x):
  likelihood_prob=np.sum(np.log(gau_pdf(k,x)))
  return likelihood_prob


def predict(x):
  y_predict=[predict2(k) for k in x]
  return y_predict

def predict2(x):
  posteriors=[]
  for k,c in enumerate(np.unique(ytrain)):
    prior_y=np.log(data_fit[2][k])
    posterior=prior_y+likelihood(k,x)
    posteriors.append(posterior)
  return np.unique(ytrain)[np.argmax(posteriors)]

def accuracy(y_true,y_pred):
  accuracy=np.sum(y_true==y_pred)/len(y_true)
  return accuracy



predict_val=predict(xtest)
print(predict_val)
acc=accuracy(ytest,predict_val)
print(acc)
```
