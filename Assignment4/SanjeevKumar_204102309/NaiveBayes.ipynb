```python
import numpy as np
import pandas as pd
import scipy.stats as st
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header = None)
print(data.head())
```

```python
#Classes
dups = data.pivot_table(index = 4, aggfunc ='size')
print(dups)
```

```python
#defining classes
classes = dups.index
class1 = data.iloc[0:50]
class2 = data.iloc[50:100]
class3 = data.iloc[100:150]
c1 = pd.DataFrame(class1)
c2 = pd.DataFrame(class2)
c3 = pd.DataFrame(class3)
#Training samples 70 percent of  total data
t1 = c1.sample(frac = 0.7)
t2 = c2.sample(frac = 0.7)
t3 = c3.sample(frac = 0.7)
#defining Gaussian function
def Gs(mu,var,x):
    f = st.norm(mu,var).pdf(x)
    return f

#testing set
def testS(c,t):
    tx = pd.DataFrame()
    for i in c.index:
        if i not in t.index:
            ty = data.loc[[i]]
            tx = pd.concat([tx,ty])
    return tx
#Necessary parameters
sLmean = [np.mean(t1.iloc[:, 0]), np.mean(t2.iloc[:, 0]), np.mean(t3.iloc[:, 0])]
sWmean = [np.mean(t1.iloc[:, 1]), np.mean(t2.iloc[:, 1]), np.mean(t3.iloc[:, 1])]
pLmean = [np.mean(t1.iloc[:, 2]), np.mean(t2.iloc[:, 2]), np.mean(t3.iloc[:, 2])]
pWmean = [np.mean(t1.iloc[:, 3]), np.mean(t2.iloc[:, 3]), np.mean(t3.iloc[:, 3])]
sLstd = [np.std(t1.iloc[:, 0]), np.std(t2.iloc[:, 0]), np.std(t3.iloc[:, 0])]
sWstd = [np.std(t1.iloc[:, 1]), np.std(t2.iloc[:, 1]), np.std(t3.iloc[:, 1])]
pLstd = [np.std(t1.iloc[:, 2]), np.std(t2.iloc[:, 2]), np.std(t3.iloc[:, 2])]
pWstd = [np.std(t1.iloc[:, 3]), np.std(t2.iloc[:, 3]), np.std(t3.iloc[:, 3])]
s = [testS(class1,t1),testS(class2,t2),testS(class3,t3)]

testingset = pd.concat(s)
##Prediction

x = np.empty([len(t1.index),3])
#convertig testing set into a data matrix only
testingsetm = testingset.to_numpy()
testingsetm = np.delete(testingsetm,4,1)
#defining prediction function
def predict(a):
    a = np.array(a)
    f = []
    for i in range(3):
        #Log being a monotonous function and prior probabilities being equal
        v = np.log(Gs(sLmean[i],sLstd[i],a[0]))+np.log(Gs(sWmean[i],sWstd[i],a[1]))+np.log(Gs(pLmean[i],pLstd[i],a[2]))+np.log(Gs(pWmean[i],pWstd[i],a[3]))
        f.append(v)
    pred = classes[f.index(max(f))]

    return pred
#initialising predicting set
predicted = []
original = list(testingset.iloc[:,4])
print('ORIGINAL LABELS',original)

```

```python
count = 0
for i in range(testingsetm.shape[0]):
    predicted.append(predict(testingsetm[i]))
    if predicted[i]==original[i]:
        count = count+1
error = (testingsetm.shape[0] - count)/testingsetm.shape[0]
print('PREDICTED LABELS',predicted)
```

```python
print('error percentage = ',error*100)
```
