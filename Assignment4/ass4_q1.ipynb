```
import numpy as np
import matplotlib.pyplot as plt
```

```
x=np.arange(-10,10,.1)
sigma=1
mean=0
x_N_0_1=(1/(2*np.pi*sigma**2)**.5)*np.exp(-(x-mean)**2/(2*sigma**2))
plt.plot(x,x_N_0_1)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title('normal distribution with N(0,1)')
plt.show()
```

```
sigma=2
mean=3
x_N_3_2=(1/(2*np.pi*sigma**2)**.5)*np.exp(-(x-mean)**2/(2*sigma**2))
plt.plot(x,x_N_3_2)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title('normal distribution with N(3,2)')
plt.show()
```

# **two dimensional normal distribution plot**

```
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
x1=np.linspace(-1,1,100)
x2=np.linspace(-1,1,100)
mean=np.array([[0],[0]])
covarience_matrix=np.array([[1,.6],[.6,2]])

x=np.array([x1,x2])
det_covarience_matrix=np.linalg.det(covarience_matrix) 

inv_covarience_matrix=np.linalg.inv(covarience_matrix)
mahalanobis_dist=np.dot(np.dot((x-mean).transpose(),inv_covarience_matrix),(x-mean))
p_x=(1/(2*np.pi*det_covarience_matrix)**.5)*np.exp(-(mahalanobis_dist)/2)

x1,x2=np.meshgrid(x1,x2)
fig=plt.figure()
ax=fig.gca(projection='3d')
ax.set_title('SURFACE PLOT OF PDF')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('pdf')
surf=ax.plot_surface(x1,x2,p_x,rstride=1,cstride=1,cmap=cm.hsv,linewidth=1)

fig.colorbar(surf,shrink=.5,aspect=10)
fig,ax=plt.subplots(1,1)

cp = ax.contourf(x1, x2,p_x)
fig.colorbar(cp) 
ax.set_title('Filled Contours Plot')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
plt.show()
```
