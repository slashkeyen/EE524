```python
from math import exp
from math import pi
from math import sqrt
def mean_find(data2):
    #FIND THE MEAN OF THE INCOMING DATASET
    temp=list()
    mean_data=dict()
    for key in data2:
        temp=data2[key]
        mean_data[key]=[]
        mean_list=[]
        temp2=[]
        for j in range(len(temp[0])-1):
            #CHANGE THE MXN DATASET TO NXM FOR SIMPLICITY TO CALCULATE THE ADDITION
            new=[]
            for i in range(len(temp)):
                new.append(temp[i][j])
            temp2.append(new)
        for i in range(len(temp2)):
            #FINDING THE MEAN
            k=0
            for j in range(len(temp2[i])):
                k+=temp2[i][j]
            mean_list.append(k/len(temp))
        mean_data[key]=mean_list
    return mean_data
def std_dev(separate,mean2):
    #CALCULATION OF THE STANDARD DEVIATION
    stddeviation=dict()
    for key in separate:
        stddeviation[key]=[]
        temp=separate[key]
        for j in range(len(temp[0])-1):
            a=0
            for i in range(len(temp)):
                a+=(temp[i][j]-mean2[key][j])**2
            stddeviation[key].append(sqrt(a))
    return stddeviation
def class_separation(data):
    #SEPARATION OF CLASS OF THE DATA PROVIDED
    separated=dict()
    for i in range (len(data)):
        row=data[i]
        class_row=row[-1]
        if (class_row not in separated):
            separated[class_row]=list()
        separated[class_row].append(row)
    return(separated)
def gaussian(test_data2,mean3,stddev3):
    #FINDING GAUSSIAN PROBABILITY USING KNOWN MEAN AND STANDARD DEVIATION OF CLASSES
    probability=1
    mu=mean3
    sigma=stddev3
    for i in range(len(mu)):
        exponent=exp(-0.5*((test_data2[i]-mu[i])/sigma[i])**2)
        init=1/(sqrt(2*pi)*sigma[i])
        probability*=(init*exponent)
        #HERE WE FIND THE TOTAL PROBABILITY OF EACH ELEMENT WITH RESPECT TO MEAN & STANDARD DEVIATION AND MULTIPLY EACH OF THEM
    no_of_classes=len(mu)
    return probability/no_of_classes
def nishan():
    file=open("iris.data")
    data=[]
    test_data=[]
    l=0
    for line in file:
        data1=file.readline()
        data2=data1.rstrip().split(",")
        for i in range(len(data2)-1):
            data2[i]=float(data2[i])
        data.append(data2)
    post_separate=class_separation(data)
    mean=mean_find(post_separate)
    stddev=std_dev(post_separate,mean)
    keys=[]
    finalprob=[]
    test_data=[[4.4,3.2,1.3,0.2],[5.0,3.5,1.6,0.6],[4.8,3.0,1.4,0.3],[5.1,3.8,1.6,0.2],[4.6,3.2,1.4,0.2],[5.3,3.7,1.5,0.2],[5.0,3.3,1.4,0.2],[5.8,2.6,4.0,1.2],[5.0,2.3,3.3,1.0],[5.6,2.7,4.2,1.3],[5.7,3.0,4.2,1.2],[5.7,2.9,4.2,1.3],[6.2,2.9,4.3,1.3],[5.1,2.5,3.0,1.1],[5.7,2.8,4.1,1.3],[6.7,3.3,5.7,2.5],[6.7,3.0,5.2,2.3],[6.3,2.5,5.0,1.9],[6.5,3.0,5.2,2.0],[6.2,3.4,5.4,2.3]]
    file2=open("test_data.data")
    print("\nThe test data is")
    for k in range(0,len(test_data),4):
               print(test_data[k]," ",test_data[k+1]," ",test_data[k+2]," ",test_data[k+3])
    prob=dict()
    for key in mean:
        prob[key]=[]
        for i in range(len(test_data)):
            prob[key].append(gaussian(test_data[i],mean[key],stddev[key]))
    print("\n") 
    for i in range(len(test_data)):
        for key in mean:
            keys.append(key)
            print (test_data[i]," is having probability ",prob[key][i]," that it is in ",key," class ")
    file.close()
nishan()
```
