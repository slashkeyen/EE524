```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
mu=np.array([[0],[0]])
sig=np.empty([2,2])
n=100
x1=x2=np.linspace(-10,10,n)
z=np.empty([n,n])
sigdet=0.0
sig=np.array([[1,0.6],[0.6,2]])
sigdet=np.linalg.det(sig)
siginv=np.linalg.inv(sig)
init=1/(2*np.pi*np.sqrt(sigdet))
for i in range(n):
    for j in range(n):
        vector=np.array([[x1[i]],[x2[j]]])-mu
        u=np.matmul(np.matmul(np.matrix.transpose(vector),siginv),vector)
        z[i][j]=init*np.exp(-0.5*u)
x1,x2=np.meshgrid(x1,x2)
fig=plt.figure()
ax=fig.gca(projection='3d')
ax.plot_surface(x1,x2,z,rstride=1,cstride=1,cmap=cm.jet,linewidth=0, antialiased=False)
cset = ax.contourf(x1, x2, z, zdir='z', offset=-0.15, cmap=cm.viridis)
for angle in range(0, 360):
    ax.view_init(30, angle)
    ax.set_zlim(-0.15,0.2)
    ax.set_zticks(np.linspace(0,0.1,5))
    ax.view_init(25, -30)
    plt.show()
    plt.pause(.001)
```
