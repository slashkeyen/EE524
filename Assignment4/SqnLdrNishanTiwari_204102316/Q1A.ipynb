```python
"""
Created on Thu Oct 22 20:24:04 2020

@author: NISHAN TIWARI
"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-10,10,100)
y1=y2=np.empty([100])
mu=0
sig=1
mu2=3
sig2=2
y1=np.exp(-0.5*((x-mu)/sig)**2)/(sig*np.sqrt(2*np.pi))
plt.plot(x,y1,'k')
plt.grid(b=None,which='both',axis='both')
y2=np.exp(-0.5*((x-mu2)/sig2)**2)/(sig2*np.sqrt(2*np.pi))
plt.plot(x,y2,'r')
'''Here the different plots have different mean and different standard deviations. However, the integration of the function from
-infinity to infinity is one. so if standard deviation is low then curve will be sharp and peak value will be high, else curve 
will be flatter and peak value will be less'''
```
