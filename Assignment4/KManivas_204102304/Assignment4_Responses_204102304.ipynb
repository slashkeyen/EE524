```python
# Question 1(a)
# To import numpy and matplotlib libraries
import numpy as np
import matplotlib.pyplot as plt


# Generating values of x in [-10,10]
x = np.linspace(-10.0,10.0,100)

# Mean,variance and standard deviation of N(0,1)
u1 = 0;var1 = 1
sig1 = np.sqrt(var1)

# Mean,variance and standard deviation of N(3,2)
u2 = 3;var2 = 2
sig2 = np.sqrt(var2)

# pdf of N(u,sig**2) = (1/2psig**2)*exp(-(x-u)**2/2sig**2)

# Calculating the pdf of N(0,1) for different values of x and storing them in vector f1 of size(100)
f1 = ((np.sqrt(2*np.pi)*sig1)**(-1))*np.exp((-1)*((x-u1)**2)/(2*sig1**2))

# Calculating the pdf of N(3,2) for different values of x and storing them in vector f2 of size(100)
f2 = ((np.sqrt(2*np.pi)*sig2)**(-1))*np.exp((-1)*((x-u2)**2)/(2*sig2**2))

# Creating a figure window
fig = plt.figure()

# Creating a subplot and using it to plot N(0,1) and N(3,2)
ax1 = fig.add_subplot(111)

# Plotting N(0,1)
ax1.plot(x,f1,'b-',label='N(0,1)')

# Plotting N(3,2) on the same plot of N(0,1)
ax1.plot(x,f2,'r-.',label='N(3,2)')

# Setting x label of the plot
ax1.set_xlabel('x values')

# Setting y label of the plot
ax1.set_ylabel('PDfs, N(0,1) and N(3,2)')

# Setting title of the plot
ax1.set_title('Plot of N(0,1) and N(3,2)')

# Adding legend to distinguish N(0,1) and N(3,2)
ax1.legend()

# To display grid
ax1.grid()

# Observations:
# 1. From the plot, it is observed that the peak value of N(0,1) is greater than that of N(3,2).
#    This is because the std. deviation of N(0,1) (i.e.,1) is less than that of N(3,2) (i.e.,sqrt(2))
#    and peak value of pdf is inversely proportional to std. deviation

# 2. It is also observed that the peaks of N(0,1) and N(3,2) occur at two different values of x.
#    This is because peak of normal distribution occurs at x=mean and the mean values for N(0,1) 
#    and N(3,2) are different
```

```python
# Question 1(b)
# To import numpy and matplotlib libraries
import numpy as np
import matplotlib.pyplot as plt

# Dimension = 2
d = 2


# No. of samples of x1 and x2
n1 = 100
n2 = 100

# Lower and upper limits of x1 and x2
a = -10.0
b = 10.0

# Generating values of x1 and x2 in [-10,10]
x1 = np.linspace(a,b,n1)
x2 = np.linspace(a,b,n2)

# Meshgrid for 3D plotting
xx1,xx2 = np.meshgrid(x1,x2)

# Mean vector as given in problem of size(2,1)
u = np.array([[0],[0]])

# Covariance matrix as given in problem of size(2,2)
sig = np.array([[1,0.6],[0.6,2]])

# Initializing the values of pdf matrix,f to be zero
f = np.zeros((n2,n1))

# Calculating pdf by looping through x1 and x2 and storing them in a matrix, f of size(n2,n1)
for i in range(len(x2)):
    for j in range(len(x1)):  
        
        # Assigning values to x1 and x2 and storing them in x vector of size(2,1)
        x = np.array([[xx1[i,j]],[xx2[i,j]]])
        
        # f1 = (2pi)**(-d/2)*det(sigma))**(-1/2)
        f1 = ((2*np.pi)**(-0.5*d))*((np.linalg.det(sig))**(-0.5)) 
        
        # f2 = (x-u).T*inv(sig)*(x-u)
        f2 = np.matmul(np.matmul((x-u).T,np.linalg.inv(sig)),(x-u))
        
        # f3 = exp(-f2)
        f3 = np.exp((-1)*f2)
        
        # f = (2pi*det(sigma))**(-d/2)*exp(-(x-u).T*inv(sig)*(x-u))
        f[i,j] = f1*f3
            

# Creating a figure window
fig = plt.figure()

# Creating 2 subplots and using 1st one for surface plot
ax1 = fig.add_subplot(211,projection='3d')

# Plotting Surface plot of f using the function plot_surface(x,y,z)
ax1.plot_surface(xx1,xx2,f)

# Setting x label of the plot
ax1.set_xlabel('x1 values')

# Setting y label of the plot
ax1.set_ylabel('x2 values')

# Setting z label of the plot
ax1.set_zlabel('N(u,sigma) in 2D')

# Setting title of the plot
ax1.set_title('Surface plot of Normal distribution in 2D')

# For displaying grid lines
ax1.grid()

# Using the 2nd subplot for contour plot
ax2 = fig.add_subplot(212)

# Plotting Contour plot of f using the function contour(x,y,z)
ax2.contour(xx1,xx2,f)

# Setting x label of the plot
ax2.set_xlabel('x1 values')

# Setting y label of the plot
ax2.set_ylabel('x2 values')

# Setting title of the plot
ax2.set_title('Contour plot of Normal distribution in 2D')

# Adjusting x limits of x-axis
ax2.set_xlim([-2.0,2.0])

# Adjusting y limits of y-axis
ax2.set_ylim([-2.5,2.5])

# For displaying grid lines
ax2.grid()
```

```python
# Question 2

# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Percentage of training samples to be considered (70%)
perc = 0.7

# Read data from web url into pandas data frame MD (Master data)
MD = pd.read_table("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",sep=',',header=None)

# Converting pandas dataframe to numpy array
MD = MD.to_numpy()

# Calculating total no. of data samples by counting the no. of rows in MD 
n = MD.shape[0]

# Initializing 3 lists to store the row indexes of selected samples of 3 classes for training
w1_ind = []
w2_ind = []
w3_ind = []

# Adding a new column to MD matrix to store the sample no. of each data sample
new_col = np.arange(1,n+1).reshape(n,1)
MD = np.append(MD,new_col,axis=1)

# Adding a new column with zeros to the MD matrix
MD = np.append(MD,np.zeros((n,1)),axis=1)


# Replacing the names of 3 classes by numbers 1,2 & 3 for calculation purpose 
for i in range(MD.shape[0]):
    if MD[i,4] == 'Iris-setosa':

# Collecting the row indices corr. to class-1 into list w1_ind       
        w1_ind.append(i)        

# Replacing the name of class1 by numbers 1 for calculation purpose
        MD[i,4] = 1
        
  
    elif MD[i,4] == 'Iris-versicolor':

# Collecting the row indices corr. to class-2 into list w2_ind         
        w2_ind.append(i)
        
# Replacing the names of class2 by number 2 for calculation purpose
        MD[i,4] = 2
        
    elif MD[i,4] == 'Iris-virginica':
        
# Collecting the row indices corr. to class-3 into list w3_ind 
        w3_ind.append(i)
        
# Replacing the names of class3 by number 3 for calculation purpose
        MD[i,4] = 3



# Dividing the Matrix MD into 3 new matrices using the list of row indices of 3 classes  
MD_w1 = MD[w1_ind]
MD_w2 = MD[w2_ind]     
MD_w3 = MD[w3_ind]

# No. of samples in class-1
n1 = MD_w1.shape[0]
# No. of samples in class-2
n2 = MD_w2.shape[0]
# No. of samples in class-3
n3 = MD_w3.shape[0]

# Calculating prior probability for each class using the frequency of their samples
p_w1 = n1/n
p_w2 = n2/n
p_w3 = n3/n

# No. of training samples to be selected for each class (i.e., 70%)
n1_tr = int(perc * n1) 
n2_tr = int(perc * n2) 
n3_tr = int(perc * n3)

# No. of testing samples to be selected for each class (i.e., 30%)
n1_te = n1 - n1_tr 
n2_te = n2 - n2_tr  
n3_te = n3 - n3_tr 

# Total no. of testing samples of all the classes
n_te = n1_te + n2_te + n3_te

# Randomly selecting 70% training samples of class-1 and storing the corr. row indices in a list
rand_sel_w1 = np.random.choice(MD_w1.shape[0], n1_tr,replace=False)

# List of row indices of testing data of class-1
rand_usel_w1 = list(set(list(range(MD_w1.shape[0])))-set(rand_sel_w1))

# Randomly selecting 70% training samples of class-2 and storing the corr. row indices in a list
rand_sel_w2 = np.random.choice(MD_w2.shape[0], n1_tr,replace=False)

# List of row indices of testing data of class-2
rand_usel_w2 = list(set(list(range(MD_w2.shape[0])))-set(rand_sel_w2))

# Randomly selecting 70% training samples of class-3 and storing the corr. row indices in a list
rand_sel_w3 = np.random.choice(MD_w3.shape[0], n1_tr,replace=False)

# List of row indices of testing data of class-3
rand_usel_w3 = list(set(list(range(MD_w3.shape[0])))-set(rand_sel_w3))

# Collecting training data of class-1 into a matrix
MD_w1_tr = MD_w1[rand_sel_w1]

# Collecting testing data of class-1 into a matrix
MD_w1_te = MD_w1[rand_usel_w1]

# Collecting training data of class-2 into a matrix
MD_w2_tr = MD_w2[rand_sel_w2]

# Collecting testing data of class-2 into a matrix
MD_w2_te = MD_w2[rand_usel_w2]

# Collecting training data of class-3 into a matrix
MD_w3_tr = MD_w3[rand_sel_w3]

# Collecting testing data of class-3 into a matrix
MD_w3_te = MD_w3[rand_usel_w3]

# Combining the testing data of 3 classes into a single matrix
MD_te = np.append(np.append(MD_w1_te,MD_w2_te,axis=0),MD_w3_te,axis=0)

# Collecting only the 1st 4 columns of each training data matrix into a 3 new matrices to calculate discriminant function
mat_w1_tr = (MD_w1_tr[:,:4]).astype(float)
mat_w2_tr = (MD_w2_tr[:,:4]).astype(float) 
mat_w3_tr = (MD_w3_tr[:,:4]).astype(float)

# Mean vector calculation of class1 training data set
u1 = np.mean(mat_w1_tr,axis=0).reshape(4,1)

# Covariance matrix calculation of class1 training data set
sig1 = np.cov(mat_w1_tr,rowvar=False)

# Determinant calculation of covariance matrix
sig1_det = np.linalg.det(sig1)

# Inverse calculation of covariance matrix
sig1_inv = np.linalg.inv(sig1)

# Mean vector calculation of class2 training data set
u2 = np.mean(mat_w2_tr,axis=0).reshape(4,1)

# Covariance matrix calculation of class2 training data set
sig2 = np.cov(mat_w2_tr,rowvar=False)

# Determinant calculation of covariance matrix
sig2_det = np.linalg.det(sig2)

# Inverse calculation of covariance matrix
sig2_inv = np.linalg.inv(sig2)

# Mean vector calculation of class3 training data set
u3 = np.mean(mat_w3_tr,axis=0).reshape(4,1)

# Covariance matrix calculation of class3 training data set
sig3 = np.cov(mat_w3_tr,rowvar=False)

# Determinant calculation of covariance matrix
sig3_det = np.linalg.det(sig3)

# Inverse calculation of covariance matrix
sig3_inv = np.linalg.inv(sig3)


g1 = np.zeros((n_te,1))
g2 = np.zeros((n_te,1))
g3 = np.zeros((n_te,1))

# General Function definition to calculate the discriminant function for each class sample considering Normal distribution
def g(y,u,sig_det,sig_inv,p_w):
    # g(x) = (-1/2)(x-u)t*(sig_inv)*(x-u)-(1/2)ln(|sig|) + ln(P(w))
    temp = ((-0.5)*np.matmul(np.matmul(((y-u).T),sig_inv),(y-u)))-((0.5)*np.log(np.absolute(sig_det))) + np.log(p_w)
    return temp

# Calculating the 3 discriminant function values for all the testing samples
for i in range(MD_te.shape[0]):
    x = MD_te[i,:4].reshape(4,1)
    g1[i,0] = g(x,u1,sig1_det,sig1_inv,p_w1)
    g2[i,0] = g(x,u2,sig2_det,sig2_inv,p_w2)
    g3[i,0] = g(x,u3,sig3_det,sig3_inv,p_w3)

# fetching the maximum values of discriminant functions for each sample test data and predicting the class to be 1,2 or 3
# Filling the column no. 6 of testing data matrix with predicted class values
    if g1[i,0] == max(g1[i,0],g2[i,0],g3[i,0]):
        MD_te[i,6] = 1
    elif g2[i,0] == max(g1[i,0],g2[i,0],g3[i,0]):
        MD_te[i,6] = 2
    elif g3[i,0] == max(g1[i,0],g2[i,0],g3[i,0]):
        MD_te[i,6] = 3

# Creating a table using numpy to show the serial no, predicted value, true values, correctness of prediction of all the testing data        
disp_table = np.zeros((n_te,4))
disp_table[:,0] = MD_te[:,5]
disp_table[:,1] = MD_te[:,6]
disp_table[:,2] = MD_te[:,4]

# Filling the column 3 with 0 or 1 based on the matching of predicted values with true values
for i in range(n_te):
    if disp_table[i,1] == disp_table[i,2]:
        disp_table[i,3] = 1
    else:
        disp_table[i,3] = 0

# Calculate the no. of matched values
correct_no = np.count_nonzero(disp_table[:,3])

# Calculate the accuracy as No. of correct predictions/total no. of testing data 
accuracy = correct_no/n_te

# Converting numpy table for display into pandas data frame and adding column names as shown below for display
disp_table_pd = pd.DataFrame(disp_table,columns=['Sample No.','Predicted class','True class','Correct/wrong'])

# Replacing the class no. with class names in the pandas table for display 
disp_table_pd['Predicted class'] = disp_table_pd['Predicted class'].replace([1.0,2.0,3.0],['Iris-setosa','Iris-versicolor','Iris-virginica'])
disp_table_pd['True class'] = disp_table_pd['True class'].replace([1.0,2.0,3.0],['Iris-setosa','Iris-versicolor','Iris-virginica'])

# Displaying the table
print('Table with predicted values and true values is:\n',disp_table_pd)

# Display the accuracy of prediction
print('The accuracy of prediction is:', accuracy*100.0,'%')

# Creating a confusion matrix and evaluating the entries
conf_mat = np.zeros((3,3),dtype='int')
for i in range(n_te):
    temp1 = int(disp_table[i,1])
    temp2 = int(disp_table[i,2])
    conf_mat[temp1-1,temp2-1] += 1

# Creating lists of row and columns labels for the confusion matrix
list1 = ['Predicted class(w1)','Predicted class(w2)','Predicted class(w3)']
list2 = ['True class(w1)','True class(w2)','True class(w3)']

# Converting conf. matrix in numpy to pandas with row and columns headers for display
conf_mat_pd = pd.DataFrame(conf_mat,index=list2,columns=list1)

# Displaying the confusion matrix
print('The confusion matrix after prediction for the given data is:\n',conf_mat_pd)

# Creating separate list for sample data serial No.'s, their predicted values and their true values
samp_list = disp_table[:,0] + 1
pred_list = disp_table[:,1]
true_list = disp_table[:,2]

# Creating a figure window
fig = plt.figure()

# Plotting bar-graph of predicted and true values as consecutive bins vs sample serial No.'s
plt.bar(samp_list,pred_list,width=0.5,label='Predicted values',color='b')
plt.bar(samp_list+0.5,true_list,width=0.5,label='True values',color='r')

# Labelling the x-axis
plt.xlabel('Serial No. of testing samples as per the given data set')


# Labelling the y-axis
plt.ylabel('Predicted and True values')


# Adding supertitle and subtitle to the bar graph
plt.suptitle('Bar chart of Predicted and true values vs test sample serial No.')
plt.title('Class No.1 --> Iris-setosa, Class No.2 --> Iris-versicolor, Class No.3 --> Iris-virginica')

# Adding legend to explaing predicted values and true values
plt.legend()

# For showing grid lines on the bar graph
plt.grid()









    






```
