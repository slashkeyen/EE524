```python
from math import sqrt
def normalize(data_sep):
    minimum=dict()
    maximum=dict()
    normal=dict()
    for key in data_sep:
        maximum[key]=[0,0,0,0]
        minimum[key]=[100,100,100,100]
        for j in range(len(data_sep[key][0])):
            for k in range(len(data_sep[key])):
                numb=data_sep[key][k][j]
                if maximum[key][j]<=numb:
                    maximum[key][j]=numb
                if minimum[key][j]>numb:
                    minimum[key][j]=numb
        normal[key]=[]
        for j in range(len(data_sep[key])):
            newlist2=[]
            for k in range(len(data_sep[key][0])):
                newlist2.append((data_sep[key][j][k]-minimum[key][k])/(maximum[key][k]-minimum[key][k]))
            normal[key].append(newlist2)
    return(normal)
def class_separation(data):
    #SEPARATION OF CLASS OF THE DATA PROVIDED
    separated=dict()
    for i in range (len(data)):
        row=data[i]
        class_row=row[-1]
        if (class_row not in separated):
            separated[class_row]=list()
        separated[class_row].append(row)
    return(separated)
def eucl_dist(list1,list2):
    dis=0
    for i in range(len(list1)):
        dis+=(list1[i]-list2[i])**2
    dis=sqrt(dis)
    return dis
def near_neib(data_set,test,neib):
    distance=dict()
    least=dict()
    closest=[]
    least5=[]
    c=dict()
    classes=[]
    num_neib=[]
    for key in data_set:
        distance[key]=[]
        least[key]=[]
        for i in range(len(data_set[key])):
            distance[key].append(eucl_dist(data_set[key][i],test))
        distance[key].sort()
        for j in range(neib):
            least[key].append(distance[key][j])
        c[key]=[]
    for key in least:
        for i in range(len(least[key])):
            closest.append(least[key][i])
    closest.sort()
    for p in range(neib):
        least5.append(closest[p])
    #print("least 5",least5)
    for n in least5:
        i=0
        for key in least:
            for p in range(len(least[key])):
                if n==least[key][p]:
                    if i==0:
                        c[key].append(n)
                        i+=1
                    elif n!=c[key][i-1]:
                        c[key].append(n)
                        i+=1
    for key in c:
        num_neib.append(len(c[key]))
        classes.append(key)
    class_data=classes[0]
    maxi=num_neib[0]
    for d in range(len(num_neib)-1):
        if maxi<int(num_neib[d+1]):
            maxi=num_neib[d+1]
            class_data=classes[d+1]
    #print(c)
    return (class_data)
def nishan():
    file=open("iris.data")
    data=[]
    csvdata=[]
    test_data=[]
    classdata=[]
    neib_min=1
    neib_max=15
    num_of_keys=0
    k_incorrect=[]
    for line in file:
        data1=file.readline()
        data2=data1.rstrip().split(",")
        for i in range(len(data2)-1):
            data2[i]=float(data2[i])
        data.append(data2)
    post_separate=class_separation(data)
    sep_data=dict()
    test_data=[[5.1, 2.5, 3.2, 4.3],[4.9, 3, 1.4, 0.2],[6, 3, 4.8, 1.8]]
    for key in post_separate:
        sep_data[key]=[]
        for i in range(len(post_separate[key])):
            newlist=[]
            for j in range(len(post_separate[key][0])-1):
                newlist.append(post_separate[key][i][j])
            sep_data[key].append(newlist)
    normalized=normalize(sep_data)
    for key in sep_data:
        num_of_keys+=1
    file.close()
    for neibour in range(neib_min,neib_max):
        for q in range(len(test_data)):
            classdata.append(near_neib(sep_data,test_data[q],neibour))
    j=neib_min
    for i in range(0,(neib_max-neib_min)*num_of_keys-2,num_of_keys):
        if classdata[i]=='Iris-versicolor':
            print("class of ",test_data[0],"using number of nearest neighbours ",j," is ",classdata[i],"and is 100% accurate")
            j+=1
        else:
            print("class of ",test_data[0],"using number of nearest neighbours ",j," is ",classdata[i],"and is innacurate")
            j+=1
            k_incorrect.append(j-1)
    j=neib_min
    print("\n")
    for i in range(1,(neib_max-neib_min)*num_of_keys-1,num_of_keys):
        if classdata[i]=='Iris-setosa':
            print("class of ",test_data[1],"using number of nearest neighbours ",j," is ",classdata[i],"and is 100% accurate")
            j+=1
        else:
            print("class of ",test_data[1],"using number of nearest neighbours ",j," is ",classdata[i],"and is innacurate")
            j+=1
            k_incorrect.append(j-1)
    j=neib_min
    print("\n")
    for i in range(2,(neib_max-neib_min)*num_of_keys,num_of_keys):
        if classdata[i]=='Iris-virginica':
            print("class of ",test_data[2],"using number of nearest neighbours ",j," is ",classdata[i],"and is 100% accurate")
            j+=1
        else:
            print("class of ",test_data[2],"using number of nearest neighbours ",j," is ",classdata[i],"and is innacurate")
            j+=1
            k_incorrect.append(j-1)
    k_incorrect.sort()
    print("\n\nthe values of k which cannot be used for k nearest neighbour method is",k_incorrect)
nishan()
```
