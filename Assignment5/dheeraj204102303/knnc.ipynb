```
import numpy as np
def ecledian_dist(row1,row2):
    dist=0
    for i in range(len(row1)):
        dist=dist+(row1[i]-row2[i])**2
    return dist**.5

def Neighbor(total_row,wrt_row,k):
    distance=list()
    for row in total_row:
        dist=ecledian_dist(wrt_row,row)
        distance.append((row,dist))
    distance.sort(key=lambda tup:tup[1])
    neighbor=list()
    for i in range(k):
        neighbor.append(distance[i][0])
    return neighbor

#



def predict_classification(train, test_row, num_neighbors):
	neighbors = Neighbor(train, test_row, num_neighbors)
	output_values = [round(row[-1]) for row in neighbors]
	prediction = max(set(output_values), key=output_values.count)
	return prediction
def predict(xtrain,x,k):
    predict_val=[predict_classification(xtrain,i,k) for i in x]
    return predict_val

def accuracy(x,y):
    acc=np.sum(x==y)/len(x)
    return acc

from sklearn.model_selection import train_test_split
from sklearn import datasets
iris=datasets.load_iris()

x=iris.data
y=iris.target
xtest=[[5.1,2.5,3.2,4.3],[4.9,3,1.4,.2],[6,3,4.8,1.8]]
predictions=predict(x,xtest,5)
print(predictions)
# print(accuracy(predictions,y))


```
