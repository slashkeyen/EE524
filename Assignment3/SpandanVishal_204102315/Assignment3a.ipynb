```
#Question1
#Rastrigin Function
import matplotlib
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import math
import matplotlib.pyplot as plt
import numpy as np

def Rastrigin(*X):
  n=len(X)
  return 10*n + sum([(x**2 - 10 * np.cos(2 * math.pi * x)) for x in X])
#for n=1
X=np.linspace(-6,6,100)
Y=Rastrigin(X)
plt.plot(X,Y,color='purple')
plt.title('2D Rastrigin Function',fontsize=20)
plt.xlabel('x1')
plt.ylabel('f(x1)')
plt.grid(True, linewidth=2)
plt.show()

#for n=2
X1=np.linspace(-6,6,100)
X2=np.linspace(-6,6,100)
X1,X2=np.meshgrid(X1,X2)
Z = Rastrigin(X1,X2)
fig = plt.figure(figsize=(12,12))
ax = fig.gca(projection='3d')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1,x2)')
ax.set_title('3D Rastrigin function',fontsize=20)
ax.plot_surface(X1, X2, Z, cmap=cm.inferno)
fig.colorbar(cm.ScalarMappable(cmap=cm.inferno), ax=ax)
plt.show()
```

```
#Question2
#Ackley's Function
import matplotlib
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import math
import matplotlib.pyplot as plt
import numpy as np
def Ackley(*X):
  n=len(X)
  return 20+math.e-20.0*np.exp(-0.2*np.sqrt(sum([(x*x/n) for x in X]))) - np.exp(sum([(np.cos(2*np.pi*x)/n) for x in X]))

#For n=1
X=np.linspace(-6,6,100)
Y=Ackley(X)
plt.plot(X,Y,color='purple')
plt.title('2D Ackley',fontsize=20)
plt.xlabel('x1')
plt.ylabel('f(x1)')
plt.grid(True,linewidth=2)
plt.show()

#For n=2
X1=np.linspace(-6,6,100)
X2=np.linspace(-6,6,100)
X1,X2=np.meshgrid(X1,X2)
Z = Ackley(X1,X2)
fig = plt.figure(figsize=(12,12))
ax = fig.gca(projection='3d')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1,x2)')
ax.set_title('3D Ackley function',fontsize=20)
ax.plot_surface(X1, X2, Z, cmap=cm.inferno)
fig.colorbar(cm.ScalarMappable(cmap=cm.inferno), ax=ax)
plt.show()
```

```
#Question3
#Sphere Function
import matplotlib
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import math
import matplotlib.pyplot as plt
import numpy as np
def Sphere(X1,X2):
  s=0
  s+=X1**2+X2**2
  return s
X1=np.linspace(-6,6,100)
X2=np.linspace(-6,6,100)
X1,X2=np.meshgrid(X1,X2)
Z=Sphere(X1,X2)
fig = plt.figure(figsize=(12,12))
ax = fig.gca(projection='3d')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1,x2)')
ax.set_title('Sphere Function',fontsize=20)
fig.colorbar(cm.ScalarMappable(cmap=cm.inferno), ax=ax)
ax.plot_surface(X1, X2, Z, cmap=cm.inferno)
plt.show()
```

```
#Question4
#Rosenbrock Function
import matplotlib
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import math
import matplotlib.pyplot as plt
import numpy as np
def Rosenbrock(X1,X2):
  s=0
  s+=100*(X1**2-X2)**2+(1-X1)**2
  return s
X1=np.linspace(-6,6,100)
X2=np.linspace(-6,6,100)
X1,X2=np.meshgrid(X1,X2)
Z=Rosenbrock(X1,X2)
fig = plt.figure(figsize=(12,12))
ax=plt.axes(projection='3d')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1,x2)')
ax.set_title('Rosebrock Function',fontsize=20)
fig.colorbar(cm.ScalarMappable(cmap=cm.inferno), ax=ax)
ax.plot_surface(X1, X2, Z,cmap=cm.inferno)
plt.show()
```

```
#Question5
#McCormick Function
import matplotlib
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import math
import matplotlib.pyplot as plt
import numpy as np
def McCormick(X1,X2):
  s=0
  s+=np.sin(X1+X2)+(X1-X2)**2-1.5*X1+2.5*X2+1
  return s
X1=np.linspace(-6,6,100)
X2=np.linspace(-6,6,100)
X1,X2=np.meshgrid(X1,X2)
Z=McCormick(X1,X2)
fig = plt.figure(figsize=(12,12))
ax=plt.axes(projection='3d')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1,x2)')
ax.set_title('McCormick Function',fontsize=20)
fig.colorbar(cm.ScalarMappable(cmap=cm.inferno), ax=ax)
ax.plot_surface(X1, X2, Z,cmap=cm.inferno)
plt.show()
```

```
#Question6
#Matyas Function
import matplotlib
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import math
import matplotlib.pyplot as plt
import numpy as np
def Matyas(X1,X2):
  s=0
  s+=0.26*(X1**2+X2**2)-0.48*X1*X2
  return s
X1=np.linspace(-6,6,100)
X2=np.linspace(-6,6,100)
X1,X2=np.meshgrid(X1,X2)
Z=Matyas(X1,X2)
fig = plt.figure(figsize=(12,12))
ax=plt.axes(projection='3d')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1,x2)')
ax.set_title('Matyas Function',fontsize=20)
fig.colorbar(cm.ScalarMappable(cmap=cm.inferno), ax=ax)
ax.plot_surface(X1, X2, Z,cmap=cm.inferno)
plt.show()
```

```
Difference:
The gradient is smooth in case of sphere,Matyas,Mccormick and Rosenbriock whereas in case of Rastrigin and Ackley the gradient 
is not a smooth one rather it is rough
The Global minima is a single one for Ackley,Rastrigin and Sphere whereas the global minima occurs for different
x1 and x2 values in case of Rosenbrock,McCormick and Matyas functions.
```
