```
# Rastrigin funtion

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 100)     
Y = np.linspace(-5, 5, 100)     

# For n=1
n=1
R1 = 10*n + (X**2 - 10 * np.cos(2 * np.pi * X))   # calculating the Rastrigin function values
 
plt.plot(X,R1,c='R')  # plotting the curve

plt.title('Rastrigin function for n=1')
plt.ylabel('f(x1)')
plt.xlabel('x1')
plt.grid(True, color='g')

plt.show()



# For n=2
n=2
X, Y = np.meshgrid(X, Y) 

R2 = 10*n + (X**2 - 10 * np.cos(2 * np.pi * X)) + (Y**2 - 10 * np.cos(2 * np.pi * Y))   # calculating the Z-axis values
 
fig = plt.figure(figsize=(16,10)) 
ax = fig.gca(projection='3d') 
surf=ax.plot_surface(X, Y, R2, rstride=1, cstride=1,  cmap=cm.coolwarm, linewidth=0.08,   antialiased=True)   

ax.set_xlabel('$x1$', rotation=-30)  # naming the axes
ax.set_ylabel('$x2$', rotation=30)
ax.set_zlabel('$f(x1,x2)$', rotation=90)

# To add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_title('Rastrigin function for n=2')  # title for the plot
plt.show()
```

```
# Ackley funtion

import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 100)     
Y = np.linspace(-5, 5, 100)     

# For n=1
n=1
R1 = -20.0*np.exp(-0.2*np.sqrt(X**2/n)) - np.exp(np.cos(2*np.pi*X)/n) + 20 + math.e  # calculating the Ackley function values
 
plt.plot(X,R1,c='R') 

plt.title('Ackley function for n=1')
plt.ylabel('f(x1)')
plt.xlabel('x1')
plt.grid(True, color='g')

plt.show()



# For n=2
n=2
X, Y = np.meshgrid(X, Y) 

# calculating the Z-axis values
R2 =-20.0*np.exp(-0.2*np.sqrt((X**2)/n)) - np.exp((np.cos(2*np.pi*X))/n)  - 20.0*np.exp(-0.2*np.sqrt((Y**2)/n)) - np.exp((np.cos(2*np.pi*Y))/n) + 20 + math.e 
 
fig = plt.figure(figsize=(16,10)) 
ax = fig.gca(projection='3d') 
surf=ax.plot_surface(X, Y, R2,  rstride=1, cstride=1,  cmap=cm.coolwarm, linewidth=0.08,   antialiased=True)   

ax.set_xlabel('$x1$', rotation=-30)  # naming the axes
ax.set_ylabel('$x2$', rotation=30)
ax.set_zlabel('$f(x1,x2)$', rotation=90)

# To add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5,)

ax.set_title('Ackley function for n=2')  # title for the plot
plt.show()
```

```
# Sphere function : 

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X=np.linspace(-5,5,100)
Y=np.linspace(-5,5,100)
X,Y=np.meshgrid(X,Y)

Z=X**2 + Y**2   # calculating the Z-axis values

fig=plt.figure(figsize=(16,10))
ax=fig.gca(projection='3d')
surf=ax.plot_surface(X,Y,Z, cstride=1, rstride=1, cmap=cm.coolwarm, linewidth=0.08, antialiased=True)

ax.set_xlabel('$x1$',rotation=-30)
ax.set_ylabel('$x2$',rotation=30)
ax.set_zlabel('$f(x1,x2)$', rotation=90)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title('Sphere function')

plt.show()
```

```
#Rosenbrock function : f (x 1 , x 2 ) = 100(x 21 − x 2 ) 2 + (1 − x 1 ) 2 )

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X=np.linspace(-5,5,100)
Y=np.linspace(-5,5,100)
X,Y=np.meshgrid(X,Y)

Z=100*np.square(X**2 - Y) + np.square(1-X)    # calculating the Z-axis values

fig=plt.figure(figsize=(16,10))
ax=fig.gca(projection='3d')
surf=ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0.08, antialiased=True )

ax.set_xlabel('$x1$',rotation=-30)
ax.set_ylabel('$x2$',rotation=30)
ax.set_zlabel('$f(x1,x2)$', rotation=90)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title('Rosenbrock function')

plt.show()
```

```
# McCormick function : f (x, y) = sin(x + y) + (x − y) 2 − 1.5x + 2.5y + 1

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X=np.linspace(-5,5,100)
Y=np.linspace(-5,5,100)
X,Y=np.meshgrid(X,Y)

Z=np.sin(X+Y) + np.square(X-Y) - 1.5*X + 2.5*Y + 1    # calculating the Z-axis values

fig=plt.figure(figsize=(16,10))
ax=fig.gca(projection='3d')
surf=ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0.08, antialiased=True)

ax.set_xlabel('$x1$',rotation=-30)
ax.set_ylabel('$x2$',rotation=30)
ax.set_zlabel('$f(x1,x2)$', rotation=90)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title('McCormick function')

plt.show()
```

```
# Matyas function : f (xy) = 0.26(x 2 + y 2 ) − 0.48xy

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X=np.linspace(-5,5,100)
Y=np.linspace(-5,5,100)
X,Y=np.meshgrid(X,Y)

Z=0.26*(X**2 + Y**2)- 0.48*np.multiply(X,Y)   # calculating the Z-axis values

fig=plt.figure(figsize=(16,10))
ax=fig.gca(projection='3d')
surf=ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0.08, antialiased=True)

ax.set_xlabel('$x1$',rotation=-30)
ax.set_ylabel('$x2$',rotation=30)
ax.set_zlabel('$f(x1,x2)$', rotation=90)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title('Matyas function')

plt.show()
```

```
# 1. The Rastrigin and Ackley's functions have multiple local minima
# 2. The sphere function is bowl shaped
# 3. The Rosenbrock function is valley shaped
# 4. The McCormick and Matyas functions are folded-plate shaped. Along one of axes, the change is drastic and the minima is easily distinguishable. But along the other axis, this change is very gradual.

# Similarity : All of them have a distinct global minima. For Matyas, Rastrigin, Ackley's and sphere it is at (0,0). For McCormick and Rosenbrock, it is non-zero.
```
