```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
f1=np.empty([100])
f2=np.empty([100])
n=int(input("Enter the value of n: "))
if n==1:
    f1=10+x**2-10*np.cos(2*np.pi*x)
    plt.plot(x,f1,'k')
else:
    x,y=np.meshgrid(x,y)
    f2 = (x**2 - 10 * np.cos(2 * np.pi * x)) + (y**2 - 10 * np.cos(2 * np.pi * y)) + 10*n
    fig = plt.figure() 
    ax = fig.gca(projection='3d') 
    ax.plot_surface(x,y,f2, rstride=1, cstride=1,cmap=cm.nipy_spectral, linewidth=0.08)    
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-20,20,200)
y=np.linspace(-20,20,200)
n=int(input("Enter n\n"))
if n==1:
    z=-20*np.exp(-0.2*x)-np.exp(np.cos(2*np.pi*x))+20+np.exp(1)
    plt.plot(x,z,'k')
else:
    x,y=np.meshgrid(x,y)
    z=-20*np.exp(-0.2*np.sqrt(0.5*(x**2+y**2)))-np.exp(0.5*(np.cos(2*np.pi*x)+np.cos(2*np.pi*y)))+20+np.exp(1)
    fig=plt.figure()
    ax=fig.gca(projection='3d')
    ax.plot_surface(x,y,z,cmap=cm.coolwarm,linewidth=0, antialiased=False)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
x,y=np.meshgrid(x,y)
z=np.sin(x+y)+(x-y)**2-1.5*x+2.5*y+1
fig=plt.figure()
ax=fig.gca(projection='3d')
ax.plot_surface(x,y,z,cmap=cm.nipy_spectral)
plt.show()
```
