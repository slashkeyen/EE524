**Answer 1.**

The **Rastrigin function** has several local minima.The function is **continuous**, **convex**, defined on **n-dimensional** space, **multimodal**, **differentiable** and **separable**.

The function is usually evaluated on the hypercube $ x_i \in [-5.12, 5.12] $ , for all $ i = 1, . . . n$.

$$ f(x_1 . . . x_n) = 10n + \sum_{i=1}^n (x_i^2 - 10cos(2\pi x_i)) $$

For n = 1

$ y = 10 + (x^2 - 10cos(2\pi x)) $ \\


For n = 2

$ Z = 20 + ((X^2 - 10cos(2\pi X)+((Y^2 - 10cos(2\pi Y))) $ where $ X=x_1$ & $ Y =x_2$


```
###############################################################################
x= np.linspace(-5.12, 5.12, 100)
# Rastrigin function for n =1
y = 10 + (x**2 - 10*np.cos(2* np.pi * x))
# Plotting 2D plot of y
fig, ax = plt.subplots()
ax.plot(x,y, 'r',  label=r"$ y = 10 + (x^2 - 10cos(2\pi x)) $")
ax.legend(loc=0)
ax.set_title('2D plot of Rastrigin function', fontsize=12)
ax.set_xlabel('$x$', fontsize=18)
ax.set_ylabel('$y$', fontsize=18)
plt.show()
###############################################################################

x = np.linspace(-5.12, 5.12, 100)
y = np.linspace(-5.12, 5.12, 100)
X, Y = np.meshgrid(x, y)
# Rastrigin function for n = 2
Z = 20+ ((X**2 - 10 * np.cos(2 * np.pi * X))+(Y**2 - 10 * np.cos(2 * np.pi * Y)))
# Plotting 3D plot Rastrigin function
fig = plt.figure(figsize=(15,10))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5)
ax.set_title('3D plot of Rastrigin function', fontsize=12)
ax.set_xlabel('$X$', fontsize=18)
ax.set_ylabel('$Y$', fontsize=18)
ax.set_zlabel('$Z$', fontsize=18)
plt.show()
```

**Answer 3.**

The **Sphere function** has d local minima except for the global one. The function is **convex**, defined on **n-dimensional** space, **differentiable**, **separable** and **unimodal**.

The function is usually evaluated on the hypercube $x_i \in [-5.12, 5.12]$, for all $ i = 1,. . , d $ .

$$ f(x_1. .  . x_2) = \sum_{i=1}^n x_i^2 $$

For n = 2

$ Z = X^2 + Y^2 $ where $ X=x_1$ & $ Y =x_2$


```
X = np.linspace(-5.12, 5.12, 100)
Y = np.linspace(-5.12, 5.12, 100)
X, Y = np.meshgrid(X, Y)
# Sphere function for n = 2 
Z = X**2 + Y**2
# Plotting 3D plot for Sphere function
fig = plt.figure(figsize=(12,9))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5)
ax.set_title('3D plot of Sphere function', fontsize=12)
ax.set_xlabel('$X$', fontsize=18)
ax.set_ylabel('$Y$', fontsize=18)
ax.set_zlabel('$Z$', fontsize=18)
plt.show()
```

**Answer 4.**

The **Rosenbrock function**, also referred to as the Valley or Banana function and it is a popular test problem for gradient-based optimization algorithms. The function is **continuous**, **convex** defined on **n-dimensional** space, **multimodal**, **differentiable** and **non-separable**.

The function is usually evaluated on the hypercube $ x_i \in [-5, 10]$ ,  for all $ i = 1, . .  d$ , although it may be restricted to the hypercube $ x_i \in [-2.048, 2.048] $, for all $i = 1,. . , d$ .

$$ f(x) = 100\sum_{i=1}^{n-1}(x_{i-1}^2-x_i)^2 + (1- x_1)^2 $$

for  n = 2

$ f(x_1, x_2) = 100(x_1^2 - x_2)^2 + (1 - x_1)^2 $ where $ X=x_1$ & $ Y =x_2$



```
X = np.linspace(-10, 10, 100)
Y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(X, Y)
# Rosenbrock function for n = 2
Z = 100*((X**2 - Y)**2)+ (1 - X)**2
# Plotting 3D plot for Rosenbrock function
fig = plt.figure(figsize=(12,9))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5)
ax.set_title('3D plot of Rosenbrock function', fontsize=12)
ax.set_xlabel('$X$', fontsize=18)
ax.set_ylabel('$Y$', fontsize=18)
ax.set_zlabel('$Z$', fontsize=18)
plt.show()
```

**Answer 5.**

**McCormick function** is **continuous**, **convex**, defined on **2-dimensional** space, **multimodal** and **differentiable**.

The function can be defined on any input domain but it is usually evaluated on $x \in [-1.5, 4]$ & $y \in [-3, 3]$
 

$$ f(x,y) = sin(x+y) + (x-y)^2 - 1.5x + 2.5y + 1 $$

```
X = np.linspace(-1.5, 4, 100)
Y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(X, Y)
# McCormick function
Z = np.sin(X+Y) + (X-Y)**2 - X*1.5 + Y*2.5 +1
# 3D plot of McCormick function
fig = plt.figure(figsize=(12,9))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5)
ax.set_title('3D plot of Rastrigin function', fontsize=12)
ax.set_xlabel('$X$', fontsize=18)
ax.set_ylabel('$Y$', fontsize=18)
ax.set_zlabel('$Z$', fontsize=18)
plt.show()
```

**Answer 6.**

**Matyas function** is **continuous**, **convex**, defined on **2-dimensional** space,  **unimodal**, **differentiable** and **non-separable**.

The function can be defined on any input domain but it is usually evaluated on $x \in [-10, 10]$ & $y \in [-10, 10]$. 


$$ f(xy) = 0.26(x^2 + y^2) - 0.48xy $$

```
X = np.linspace(-10, 10, 100)
Y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(X, Y)
# Matyas function
Z = ((X**2 + Y**2)* 0.26) - X*Y*0.48
# 3D plot of Matyas function
fig = plt.figure(figsize=(12,9))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5)
ax.set_title('3D plot of Rastrigin function', fontsize=12)
ax.set_xlabel('$X$', fontsize=18)
ax.set_ylabel('$Y$', fontsize=18)
ax.set_zlabel('$Z$', fontsize=18)
plt.show()
```
