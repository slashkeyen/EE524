# McCormick function 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-5, 5, 100)
y=np.linspace(-5, 5, 100)
x,y = np.meshgrid(x,y)
z=( 1 + np.sin(x+y) + (x-y)**2 - 1.5*x + 2.5*y )
fig = plt.figure(figsize=(10,10))
ax =plt.axes(projection='3d')
plt.title('McCormick function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
ax.plot_surface(x,y,z,cmap='viridis')
plt.show()
```

# Matyas function 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-5, 5, 100)
y=np.linspace(-5, 5, 100)
x,y = np.meshgrid(x,y)
z = 0.26 * (x**2 +y**2) - 0.48*x*y # z is a f(x,y)
fig = plt.figure(figsize=(10,10))
ax =plt.axes(projection='3d')
plt.title('Matyas function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
ax.plot_surface(x,y,z,cmap='viridis')
plt.show()
```

# Rosenbrock function 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
 
x1=np.linspace(-5, 5, 100)
x2=np.linspace(-5, 5, 100)
x1,x2 = np.meshgrid(x1,x2)
z = (100*(x1**2 - x2)**2 +(1-x1)**2) # z is f(x1,x2)
fig = plt.figure(figsize=(10,10))
ax =plt.axes(projection='3d')
plt.title('Rosenbrock function')
ax.set_xlabel('x1')
ax.set_xlabel('x2')
ax.set_zlabel('f(x1,x2)')
ax.plot_surface(x1,x2,z,cmap='viridis')
plt.show()
```

# Sphere function

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
 
x1=np.linspace(-5, 5, 100)
x2=np.linspace(-5, 5, 100)
x1,x2 = np.meshgrid(x1,x2)
z= x1**2 + x2**2
fig = plt.figure(figsize=(10,10))
ax =plt.axes(projection='3d')
plt.title('Sphere function')
ax.set_xlabel('x1')
ax.set_xlabel('x2')
ax.set_zlabel('f(x1,x2)')
ax.plot_surface(x1,x2,z,cmap='viridis')
plt.show()
```

# Rastrigin function 

```python
#rastrigin function for n= 1
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(-10,10,200)
n=1
f= 10*np.ones(x.shape)+(x**2 + 10*np.cos(2*np.pi*x))
figure = plt.figure(figsize=(5,5))
plt.plot(x,f,'r')
plt.title("rastrigin function for n = 1")
plt.xlabel('x1')
plt.ylabel('f(x1)')
plt.grid(True)

#rastrigin function for n = 2

import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm


x1 = np.linspace(-5,5,100)
x2 = np.linspace(-5,5,100)
x1,x2 = np.meshgrid(x1,x2)
z = 20+ (x1**2 + 10*np.cos(2*np.pi*x1))+(x2**2 +10*np.cos(2*np.pi*x2)) #z is f(x1,x2)
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
plt.title('rastrigin function for n=2')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1,x2)')
ax.plot_surface(x1,x2,z,cmap='viridis')
plt.show()
```

# Ackley's function

```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
x1 = np.linspace(-10,10,200)
x2 = np.linspace(-10,10,200)
x1,x2=np.meshgrid(x1,x2)
sum1 = -20 *np.exp(-0.2*np.sqrt(x1**2+x2**2)/2)
sum2 = -np.exp((np.cos(2*np.pi*x1)+np.cos(2*np.pi*x2))/2)
f1= sum1 +sum2 +20 +np.exp(1) # z is f(x1,x2)
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
plt.title('ackley function for n=2')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1,x2)')
ax.plot_surface(x1,x2,f1,cmap='viridis')
plt.show()


# Ackley's function for n =1
import numpy as np
import math 
import matplotlib.pyplot as plt
x=np.linspace(-10,10,200)
f = 20 + np.exp(1)-20*np.exp(-0.2*np.sqrt(x**2)/1)-np.exp(np.cos(2*np.pi*x)/1)
figure = plt.figure(figsize=(5,5))
plt.plot(x,f,'r')
plt.title("ackley function for n = 1")
plt.xlabel('x1')
plt.ylabel('f(x1)')
plt.grid(True)

```
