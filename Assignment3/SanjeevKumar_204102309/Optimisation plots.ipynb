```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
#n = 1
x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)
func = 10+x1**2-10*np.cos(2*np.pi*x1)
plt.title('Rastrigin function with one variable')
plt.grid()
plt.plot(x1,func,color = 'c')
plt.show()
```

```python
#n =2
x1,x2 = np.meshgrid(x1,x2)
func2 = 20 + x1**2-10*np.cos(2*np.pi*x1)+x2**2-10*np.cos(2*np.pi*x2)
ax = plt.axes(projection = '3d')
ax.plot_surface(x1,x2,func2,color = 'b')
plt.title('Rastrigin function with two variable')
plt.grid()
plt.show()
```

```python
x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)
#n =1
func = -20*np.e**(-0.2*np.sqrt(x1**2))-np.e**(np.cos(2*np.pi*x1))+20+np.e
plt.plot(x1,func,color = 'c')
plt.title("Ackley's function with one variable")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)
x1,x2 = np.meshgrid(x1,x2)
t = np.sqrt(0.5*(x1**2+x2**2))
u = np.cos(2*np.pi*x1) + np.cos(2*np.pi*x2)
y = -20*np.e**(-0.2*t)-np.e**(0.5*u)+20+np.e
z = np.array(y)
ax=plt.axes(projection = '3d')
plt.grid()
plt.title("Ackley's function with two variable")
ax.plot_surface(x1,x2,y)
plt.show()
```

```python
func2 = 0.5*(x1**2+x2**2)
ax = plt.axes(projection = '3d')
ax.plot_surface(x1,x2,func2)
plt.title('Sphere function')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)
x1,x2 = np.meshgrid(x1,x2)
func2 = 100*(((x1**2-x2)**2) + (1-x1**2))
ax = plt.axes(projection = '3d')
ax.plot_surface(x1,x2,func2)
plt.title('Rosenbrock function')
plt.grid()
plt.show()
```

```python
x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)
x1,x2 = np.meshgrid(x1,x2)
func2 = np.sin(x1+x2)+(x1-x2)**2-1.5*x1+2.5*x2+1
ax = plt.axes(projection = '3d')
ax.plot_surface(x1,x2,func2)
plt.title('McCormick function')
plt.grid()
plt.show()
print(func2.min())
```

```python
func2 = 0.26*(x1**2+x2**2)-0.48*x1*x2
ax = plt.axes(projection = '3d')
ax.plot_surface(x1,x2,func2)
plt.title('Matayas function')
plt.grid()
plt.show()
```

From the above optimisation curves. It can be well observed that for all the functions we have the global minima at the origin,except the McCormick function.
