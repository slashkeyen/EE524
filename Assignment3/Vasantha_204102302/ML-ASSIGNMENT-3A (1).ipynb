# QUESTION-1

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

#2-D PLOT FOR RASTRIGIN FUNCTION AT n = 1

n1 = 1
x1 = np.linspace(-5,5,100)
y1 = 10*n1+(x1**2 - 10 * np.cos(2 * np.pi * x1))
plt.plot(x1,y1,c='red')
plt.xlabel('x1-axis')
plt.ylabel('y1-axis')
plt.title('2-D PLOT FOR RASTRIGIN  FUNCTION')
plt.show()

#3-D PLOT FOR RASTRIGIN FUNCTION AT n = 2

n2 = 2
x1 = np.linspace(-5,5,100) 
x2 = np.linspace(-5,5,100) 
x1, x2 = np.meshgrid(x1, x2)
z =10*n2 + (x1**2 -10 * np.cos(2*np.pi * x1)) + (x2**2 -10 * np.cos(2*np.pi * x2))
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_title('3-D PLOT FOR RASTRIGIN FUNCTION')
ax.set_xlabel('x1-axis')
ax.set_ylabel('x2-axis')
ax.set_zlabel('z-axis')
surf = ax.plot_surface(x1, x2, z,rstride=1, cstride=1, cmap=cm.RdPu,linewidth=1)  
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
```

# QUESTION-2

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

#2-D PLOT FOR ACKLEYS FUNCTION AT n = 1

x = np.linspace(-5,5,200)
z1 =  -20*np.exp(-0.2*np.sqrt(x**2/1)) -np.exp(np.cos(2*np.pi*x)/1) + np.exp(1)+ 20
plt.plot(x,z1,c='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2-D PLOT FOR ACKLEYS FUNCTION')
plt.show()

#3-D PLOT FOR ACKLEYS FUNCTION AT n = 2

x1 = np.linspace(-5,5,200)
x2 = np.linspace(-5,5,200)
x1,x2 = np.meshgrid(x1,x2)
z2=0.2 + np.exp(1) + -20 * np.exp(-0.2 * np.sqrt(x1*x1 + x2*x2) / 2)+ -np.exp((np.cos(2*np.pi*x1) + np.cos(2*np.pi*x2)) / 2)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_title('3-D PLOT FOR ACKLEYS FUNCTION')
ax.set_xlabel('x1-axis')
ax.set_ylabel('x2-axis')
ax.set_zlabel('z2-axis')
surf = ax.plot_surface(x1,x2,z2,rstride=1, cstride=1, cmap=cm.gist_ncar,linewidth=1)  
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
          
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

x1 = np.linspace(-10,10,100) 
x2 = np.linspace(-10,10,100) 
x1, x2 = np.meshgrid(x1, x2)
Z = x1**2 +x2**2


fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_title('PLOT FOR SPHERE  FUNCTION')
ax.set_xlabel('x1-axis')
ax.set_ylabel('x2-axis')
ax.set_zlabel('Z-axis')
surf = ax.plot_surface(x1,x2,Z,rstride=1, cstride=1, cmap=cm.jet,linewidth=1)  
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
          
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

x1 = np.linspace(-1,1,100) 
x2 = np.linspace(-1,1,100) 
x1, x2 = np.meshgrid(x1, x2)
Z = 100*(x1**2-x2)**2 +(1-x1)**2

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_title('PLOT FOR ROSENBROCK FUNCTION')
ax.set_xlabel('x1-axis')
ax.set_ylabel('x2-axis')
ax.set_zlabel('Z-axis')
surf = ax.plot_surface(x1, x2, Z,rstride=1, cstride=1, cmap=cm.hsv,linewidth=1)  
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np


x = np.linspace(-10,10,100) 
y = np.linspace(-10,10,100) 
x,y = np.meshgrid(x,y)
z = np.sin(x+y)+(x-y)**2-1.5*x+2.5*y+1

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_title('PLOT FOR MCCORMICK FUNCTION')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
surf = ax.plot_surface(x, y, z,rstride=1, cstride=1, cmap=cm.jet,linewidth=1)  
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
```

# QUESTION-6

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

x = np.linspace(-10,10,100) 
y = np.linspace(-10,10,100) 
x,y = np.meshgrid(x,y)
Z = 0.26*(x**2+y**2) - (0.48*x*y)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_title('PLOT FOR Matyas FUNCTION')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('Z-axis')
surf = ax.plot_surface(x, y, Z,rstride=1, cstride=1, cmap=cm.spring,linewidth=1)  
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
```

```python
# THE DIFFERENCES I OBSERVED IN THE ABOVE PLOTS
1.RASTRIGIN FUNCTION---# MULTIMODAL,,,FLATTER THAN ACKLEYS FUNCION PLOT,,,IT HAS SEVERAL LOCAL MINIMA WHICH ARE REGULARLY DISTRIBUTED
2.ACKLEYS FUNCTION-----# MULTIMODAL,,,LARGE FUNNEL SHAPED,,,MANY LOCAL MINIMA AND CHARACTERISED BY HOLE AT CENTRE AND FLAT SURFACE AT OUTER REGION
3.MCCORMICK FUNCTION---# MULTIMODAL,,,PLATE SHAPED,,,IT HAS NO LOCAL MINIMA EXCEPT THE GLOBAL ONE
4.SPHERE FUNCTION------# UNIMODAL,,,BOWL SHAPED,,,IT HAS N LOCAL MINIMA
5.MATYAS FUNCTION------# UNIMODAL,,,CURVE PLATEAU SHAPE,,,IT HAS NO LOCAL MINIMA EXCEPT THE GLOBAL ONE
6.ROSENBORCK FUNCTION--# UNIMODAL,,,VALLEY SHAPED,,,GLOBAL MINIMUM LIES IN A NARROW PARABOLIC VALLEY
```
