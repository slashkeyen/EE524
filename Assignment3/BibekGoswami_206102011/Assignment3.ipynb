```
import numpy as np
from matplotlib import pyplot as plt
import math
```

```
#Rasrigin function

#n=1

x = np.arange(0.0, 5.0, 0.1)
f = []

for i in range(len(x)):
  temp = 10*1 + ((x[i]*x[i])-np.cos(2*np.pi*x[i]))
  f.append(temp)

plt.plot(x,f)
plt.show()

#n=2

x1 = np.arange(0.0, 5.0, 0.1)
x = np.array(list(zip(x1,x1)))

f = []

for j in range(2):
  f_1 = []
  for i in range(len(x1)):
    temp = 10*2 + ((x[i][j]*x[i][j])-np.cos(2*np.pi*x[i][j]))
    f_1.append(temp)
  f.append(f_1)

f=np.array(f)

n = [1,2]

X1, N = np.meshgrid(x1, n)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, N, f,500)

plt.show()
```

```
#Ackley's function

#n=1

x = np.arange(0.0, 5.0, 0.1)
f = []

for i in range(len(x)):
  temp = -20*np.exp(math.sqrt((1/2)*x[i]**2))-np.exp((1/2)*np.cos(2*np.pi*x[i]))+20+1
  f.append(temp)

plt.plot(x,f)
plt.show()

#n=2,e=1

x1 = np.arange(0.0, 5.0, 0.1)
x = np.array(list(zip(x1,x1)))

f = []

for j in range(2):
  f_1 = []
  for i in range(len(x1)):
    temp = -20*np.exp(math.sqrt((1/2)*x[i][j]**2))-np.exp((1/2)*np.cos(2*np.pi*x[i][j]))+20+1
    f_1.append(temp)
  f.append(f_1)

f=np.array(f)

n = [1,2]

X1, N = np.meshgrid(x1, n)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, N, f,500)

plt.show()
```

```
#Sphere function
#n=2

x1 = np.arange(0.0, 5.0, )
x = np.array(list(zip(x1,x1)))

f = []

for j in range(2):
  f_1 = []
  for i in range(len(x1)):
    temp = x[i][j]*x[i][j]
    f_1.append(temp)
  f.append(f_1)

f=np.array(f)

n = [0,1]

X1, N = np.meshgrid(x1, n)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, N, f,500)

plt.show()
```

```
#Rosenbrock function

x1 = np.arange(0.0, 5.0, 0.1)
x2 = np.arange(5.0, 10.0, 0.1)

X1,X2 = np.meshgrid(x1,x2)
f = 100*((X1**2-X2)**2+(1-X1)**2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, X2, f,500)

plt.show()
```

```
#McCormick function

x = np.arange(0.0, 5.0, 0.1)
y = np.arange(0.0, 5.0, 0.1)

X, Y = np.meshgrid(x,y)

f = np.sin(X+Y)+((X-Y)**2)-(1.5*X)+(2.5*Y)+1

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, f,500)

plt.show()
```

```
#Matyas function

x = np.arange(0.0, 5.0, 0.1)
y = np.arange(0.0, 5.0, 0.1)

X,Y = np.meshgrid(x,y)

f = 0.26*(X**2+Y**2) - (0.48*X*Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, f,500)

plt.show()
```
