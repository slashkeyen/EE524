```python
# Question no.-1
# Rastrigin Function
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X1 = np.linspace(-5, 5, 100)     
X2 = np.linspace(-5, 5, 100)     
 
#for n=1
Z1 = (X1**2 - 10 * np.cos(2 * np.pi * X1)) + 10
fig = plt.figure() 
plt.plot(X1,Z1)
plt.xlabel('X1----->')
plt.ylabel('Z1------>')
plt.title('plot for n = 1')
plt.show()

#for n= 2

X1, X2 = np.meshgrid(X1, X2)
Z2 = (X1**2 - 10 * np.cos(2 * np.pi * X1)) + \
  (X2**2 - 10 * np.cos(2 * np.pi * X2)) + 20

 
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X1, X2, Z2, rstride=1, cstride=1,
  cmap=cm.gist_rainbow, linewidth=0.1,
  antialiased=True) 
ax.set_xlabel('X1----->')
ax.set_ylabel('X2------>')
ax.set_zlabel('Z2----->')
plt.title('plot for n = 2 Rastrigin Function ')
plt.show()
```

```python
# Question No.-2
# Ackley's Function
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D


X1 = np.linspace(-5, 5, 100)     
X2 = np.linspace(-5, 5, 100)

#for n=1
Z1 = -20*np.exp(-0.2*np.sqrt(X1**2)) \
     -np.exp(np.cos(2*np.pi*X1)) +20 + np.exp(1)
plt.xlabel('X1----->')
plt.ylabel('Z1------>')
plt.title('plot for n = 1')
plt.plot(X1,Z1)

#for n= 2

X1, X2 = np.meshgrid(X1, X2)

Z2 = -20*np.exp(-0.2*np.sqrt(0.5*(X1**2+X2**2))) \
   -np.exp(0.5*(np.cos(2*np.pi*X1)+ np.cos(2*np.pi*X2))) + 20 + np.exp(1)


fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X1, X2, Z2, rstride=1, cstride=1,
  cmap=cm.gist_rainbow, linewidth=0.1,
  antialiased=True)
ax.set_xlabel('X1----->')
ax.set_ylabel('X2------>')
ax.set_zlabel('Z2----->')
plt.title("plot for n = 2 Ackley's Function ")
plt.show()
```

```python
# Question No -3
# Sphere Function
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X1 = np.linspace(-5, 5, 100)     
X2 = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(X1, X2)
Z = (X1**2 + X2**2)
fig = plt.figure() 
ax = fig.gca(projection='3d')
ax.plot_surface(X1, X2, Z, rstride=1, cstride=1,
  cmap=cm.gist_rainbow, linewidth=0.1,
  antialiased=True)

ax.set_xlabel('X1----->')
ax.set_ylabel('X2------>')
ax.set_zlabel('Z----->')
plt.title("plot for Sphere Function ")
plt.show()
```

```python
# Question No -4
# Rosenbrock Function
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X1 = np.linspace(-2, 2, 200)     
X2 = np.linspace(-2, 2, 200)
X1, X2 = np.meshgrid(X1, X2)

Z = 100*(X1**2 - X2**2)**2 + (1-X1)**2

fig = plt.figure() 
ax = fig.gca(projection='3d')
ax.plot_surface(X1, X2, Z, rstride=1, cstride=1,
  cmap=cm.gist_rainbow, linewidth=0.1,
  antialiased=True)
ax.set_xlabel('X1----->')
ax.set_ylabel('X2------>')
ax.set_zlabel('Z----->')
plt.title("plot for Rosenbrock Function ")
plt.show()
```

```python
# Question No.-5
# McCormick Function
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 100)     
Y = np.linspace(-5, 5, 100)
X,Y = np.meshgrid(X,Y)

Z= np.sin(X+Y) + (X-Y)**2 - 1.5*X + 2.5*Y +1

fig = plt.figure() 
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
  cmap=cm.gist_rainbow, linewidth=0.1,
  antialiased=True)
ax.set_xlabel('X----->')
ax.set_ylabel('Y------>')
ax.set_zlabel('Z----->')
plt.title("plot for McCormick Function ")
plt.show()
```

```python
# Question No-6
# Matyas Function
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 100)     
Y = np.linspace(-5, 5, 100)
X,Y = np.meshgrid(X,Y)

Z = 0.26*(X**2+Y**2)-0.48*X*Y


fig = plt.figure() 
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
  cmap=cm.gist_rainbow, linewidth=0.1,
  antialiased=True)

ax.set_xlabel('X----->')
ax.set_ylabel('Y------>')
ax.set_zlabel('Z----->')
plt.title("plot for Matyas Function ")
plt.show()
```
