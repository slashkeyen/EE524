Q1. Put n = 1 and 2. This is the famous Rastrigin function. You will get a 2D plot for n = 1 and a 3D plot for n = 2.

f(x1 ... xn) = 10n + E n, i=1 (xi^2 - 10cos(2*pi*xi))

```python
#for n = 1, we have the following plot for Rastrigin function.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
   
X = np.linspace(-5, 5, 100)     
Y = np.linspace(-5, 5, 100)     
X, Y = np.meshgrid(X, Y) 

Z = 10 + (X**2 - 10 * np.cos(2 * np.pi * X)) #here n = 1, so 10*n = 10. and x1 = X
 
fig = plt.figure(figsize=(8.5,5)) #we defined figsize as 8.5, 5 
ax = fig.gca(projection='3d') 
surf = ax.plot_surface(X, Y, Z, cmap=cm.rainbow, linewidth=0, antialiased=False) #here rainbow is the color of the 3d surface. cm is actually imported from matplotlib
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('(x-axis)')
ax.set_ylabel('(y-axis)')
ax.set_zlabel('(z-axis)')    
plt.show()
```

```python
#for n = 2, we have the following plot for Rastrigin function 

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 200)     
Y = np.linspace(-5, 5, 200)     
X, Y = np.meshgrid(X, Y) 

Z = 20 + (X**2 - 10 * np.cos(2 * np.pi * X)) + (Y**2 - 10 * np.cos(2 * np.pi * Y))         #n=2, so 10*n = 20, again for n = 2, we have two terms in the summation x1 & x2. Here x1 = X & x2 = Y
 
fig = plt.figure(figsize=(8,5)) 
ax = fig.gca(projection='3d') 
surf = ax.plot_surface(X, Y, Z, cmap=cm.Spectral, linewidth=0, antialiased=False) #here Spectral is the color of the 3d surface. cm is actually imported from matplotlib
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('(x-axis)')
ax.set_ylabel('(y-axis)')
ax.set_zlabel('(z-axis)')

plt.show()
```

Q2. Plot the above function for n = 1 and 2. This is known as Ackley's function.
f(x0 ... xn) = -20 exp(-0.2*sq.root(1/n*E i=1, n xi**2) - exp(1/n*E i=1, n cos(2*pi*xi)) + 20 + e

```python
#for n = 1, we have the following plot for Ackley's function.


import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 200)     
Y = np.linspace(-5, 5, 200)     
X, Y = np.meshgrid(X, Y) 


Z = -20 * np.exp(-0.2 * np.sqrt(X**2)) - np.exp(np.cos(2 * np.pi *X)) + 20 + np.exp(1)     #here n = 1, so we have only one term under summation x1 = X
                                                
fig = plt.figure(figsize=(8,5))
ax = fig.gca(projection='3d') 
surf = ax.plot_surface(X, Y, Z, cmap=cm.gist_earth, linewidth=0, antialiased=False) #here gist_earth is the color of the 3d surface. cm is actually imported from matplotlib
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('(x-axis)')
ax.set_ylabel('(y-axis)')
ax.set_zlabel('(z-axis)')
plt.show()
```

```python
#for n = 2, we have the following plot for Ackley's function.


import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 200)     
Y = np.linspace(-5, 5, 200)     
X, Y = np.meshgrid(X, Y) 


Z = -20 * np.exp(-0.2 * np.sqrt(1/2 * (X**2 + Y**2))) - np.exp(1/2 * (np.cos(2 * np.pi *X) + np.cos(2 * np.pi * Y))) + 20 + np.exp(1)     

#here n = 2, so we have two terms under summation. x1 = X & x2 = Y


fig = plt.figure(figsize=(8,5))
ax = fig.gca(projection='3d') 
surf = ax.plot_surface(X, Y, Z, cmap=cm.gist_ncar, linewidth=0, antialiased=False) #here gist_ncar is the color of the 3d surface. cm is actually imported from matplotlib
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('(x-axis)')
ax.set_ylabel('(y-axis)')
ax.set_zlabel('(z-axis)')
plt.show()
```

Q3. Put n = 2 and plot the function. This is known as the Sphere function.
f(x1 ... xn) = E i=1, n xi**2

```python
#for n = 2, we have the following plot for Sphere function.


import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 200)     
Y = np.linspace(-5, 5, 200)     
X, Y = np.meshgrid(X, Y) 


Z = X**2 + Y**2     

#here n = 2, so we have two terms under summation. x1 = X & x2 = Y


fig = plt.figure(figsize=(8,5))
ax = fig.gca(projection='3d') 
surf = ax.plot_surface(X, Y, Z, cmap=cm.ocean, linewidth=0, antialiased=False) #here ocean is the color of the 3d surface. cm is actually imported from matplotlib
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('(x-axis)')
ax.set_ylabel('(y-axis)')
ax.set_zlabel('(z-axis)')
plt.show()
```

Q4. Plot the Rosenbrock function.
f(x1, x2) = 100((x1^2 - x2^2)^2 + (1 - x1)^2)

```python
#we have the following plot for Rosenbrock function.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 200)     
Y = np.linspace(-5, 5, 200)     
X, Y = np.meshgrid(X, Y) 


Z = 100 * (((X**2 - Y**2)**2) + ((1 - X)**2))   


fig = plt.figure(figsize=(8,5))
ax = fig.gca(projection='3d') 
surf = ax.plot_surface(X, Y, Z, cmap=cm.rainbow, linewidth=0, antialiased=False) #here ocean is the color of the 3d surface. cm is actually imported from matplotlib
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('(x-axis)')
ax.set_ylabel('(y-axis)')
ax.set_zlabel('(z-axis)')
plt.show()
```

Q5. Plot McCormick function.
f(x, y) = sin(x + y) + (x - y)^2 - 1.5*x + 2.5*y + 1

```python
#we have the following plot for McCormick function.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 200)     
Y = np.linspace(-5, 5, 200)     
X, Y = np.meshgrid(X, Y) 


Z = np.sin(X + Y) + (X + Y)**2 - 1.5*X + 2.5*Y + 1


fig = plt.figure(figsize=(8,5))
ax = fig.gca(projection='3d') 
surf = ax.plot_surface(X, Y, Z, cmap=cm.hsv, linewidth=0, antialiased=False) #here hsv is the color of the 3d surface. cm is actually imported from matplotlib
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('(x-axis)')
ax.set_ylabel('(y-axis)')
ax.set_zlabel('(z-axis)')
plt.show()
```

Q6. Plot Matyas function.
f(xy) = 0.26(x^2 + y^2) - 0.48*x*y

```python
#we have the following plot for Matyas function.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 200)     
Y = np.linspace(-5, 5, 200)     
X, Y = np.meshgrid(X, Y) 


Z = 0.26 * (X**2 + Y**2) - 0.48 * X * Y


fig = plt.figure(figsize=(8,5))
ax = fig.gca(projection='3d') 
surf = ax.plot_surface(X, Y, Z, cmap=cm.cividis, linewidth=0, antialiased=False) #here cividis is the color of the 3d surface. cm is actually imported from matplotlib
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('(x-axis)')
ax.set_ylabel('(y-axis)')
ax.set_zlabel('(z-axis)')
plt.show()
```
