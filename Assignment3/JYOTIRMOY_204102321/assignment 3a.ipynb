1.Put n = 1 and 2. This is the famous Rastrigin function. You will get a 2D plot for n = 1 and a 3D plot for n = 2.
                     f (x1 • • • xn) = 10n +  Σ  (x** 2 − 10cos(2πxi))
                                              i=1

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X1 = np.linspace(-5, 5, 100) 
X2 = np.linspace(-5, 5, 100)

Y = (X1**2 - 10 * np.cos(2 * np.pi * X1)) + 20
 
plt.plot(X1,Y)  


X1, X2 = np.meshgrid(X1, X2)
Z = (X1**2 - 10 * np.cos(2 * np.pi * X1)) + (X2**2 - 10 * np.cos(2 * np.pi * X2)) + 20
 
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X1, X2, Z, rstride=1, cstride=1,cmap=cm.gist_heat, linewidth=0.08)    
plt.show()
```

3.Put n = 2 and plot the function. This is known as the Sphere function.
   f(x1 · · · xn) =  Σ(xi)** 2

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X1 = np.linspace(-5, 5, 100)     
X2 = np.linspace(-5, 5, 100)     
X, X2 = np.meshgrid(X1,X2) 

Z = (X1**2) + (X2**2)
  
 
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X1,X2, Z, rstride=1, cstride=1,cmap=cm.nipy_spectral, linewidth=0.08)    

plt.show()
```

4.Plot the Rosenbrock function.
         f(x1, x2) = 100((x2** 1 − x2) ** 2) + (1 − x1)** 2

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X1 = np.linspace(-5, 5, 100)     
X2 = np.linspace(-5, 5, 100)     
X1, X2 = np.meshgrid(X1, X2) 

Z = 100*(X1**2-X2)**2 + ((1-X1)**2)
  
 
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X1,X2, Z, rstride=1, cstride=1,cmap=cm.nipy_spectral, linewidth=0.08)    

plt.show()
```

5.Plot McCormick function.
             f(x, y) = sin(x + y) + (x − y)** 2 − 1.5x + 2.5y + 1

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 100)     
Y = np.linspace(-5, 5, 100)     
X, Y = np.meshgrid(X, Y) 

Z = np.sin(X+Y) + (X-Y)**2 - 1.5*X + 2.5*Y + 1
  
 
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap=cm.nipy_spectral, linewidth=0.08)    

plt.show()
```

6.Plot Matyas function.
               f(xy) = 0.26(x** 2 + y** 2) − 0.48xy

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 100)     
Y = np.linspace(-5, 5, 100)     
X, Y = np.meshgrid(X, Y) 

Z = 0.26*(X**2 + Y**2) - 0.48*X*Y
  
 
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap=cm.nipy_spectral, linewidth=0.08)    

plt.show()
```

Do you see any differences between the plots? If yes, what is the main
difference?

The major difference amongst the plots is their nature of covexity.
