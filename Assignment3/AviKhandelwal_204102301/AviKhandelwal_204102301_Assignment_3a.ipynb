```python
#Rastrigin function, n=1
from math import pi
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,300)

y =  10 + x**2 - 10*np.cos(2*pi*x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Rastrigin function, n=1")
plt.show()
```

```python
#Rastrigin function, n=2
from math import pi
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')

x1 = np.linspace(-5,5,300)
x2 = np.linspace(-5,5,300)
x1, x2 = np.meshgrid(x1, x2)

z = 20 + (x1**2) + (x2**2) - 10*np.cos(2*pi*x1) - 10*np.cos(2*pi*x2)


surf = ax.plot_surface(x1,x2,z, rstride=1, cstride=1, cmap=cm.jet)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel("f(x1,x2)")
plt.title("Rastrigin function, n=2")

plt.show()
```

```python
#Ackley’s Function, n=1
from math import pi
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5,5,300)
y = -20 * np.exp((-0.2)*np.sqrt(x**2)) -  np.exp(np.cos(2*pi*x)) + 20 + np.e
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Ackley’s Function, n=1")
plt.show()
```

```python
#Ackley’s Function, n=2
from math import pi
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')

x1 = np.linspace(-5,5,300)
x2 = np.linspace(-5,5,300)
x1, x2 = np.meshgrid(x1, x2)
z = (-20)*np.exp((-0.2)*np.sqrt((x1**2 + x2**2)/2)) -  np.exp((np.cos(2*pi*x1) + np.cos(2*pi*x2))/2) + 20 + np.e
surf = ax.plot_surface(x1,x2,z, rstride=1, cstride=1, cmap=cm.jet)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel("f(x1,x2)")
plt.title("Ackley’s Function, n=2")

plt.show()
```

```python
#Sphere Function
from math import pi
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')

x1 = np.linspace(-5,5,300)
x2 = np.linspace(-5,5,300)
x1, x2 = np.meshgrid(x1, x2)

z = x1**2 + x2**2

surf = ax.plot_surface(x1,x2,z, rstride=1, cstride=1, cmap=cm.jet)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel("f(x1,x2)")
plt.title("Sphere Function")

plt.show()
```

```python
#Rosenbrock function
from math import pi
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')

x1 = np.linspace(-5,5,300)
x2 = np.linspace(-5,5,300)
x1, x2 = np.meshgrid(x1, x2)
z = 100*((x1**2 - x2)**2) + (1-x1)**2
surf = ax.plot_surface(x1,x2,z, rstride=1, cstride=1, cmap=cm.jet)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel("f(x1,x2)")
plt.title("Rosenbrock function")

plt.show()
```

```python
#McCormick function
from math import pi
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')

x1 = np.linspace(-5,5,300)
x2 = np.linspace(-5,5,300)
x1, x2 = np.meshgrid(x1, x2)
z = np.sin(x1+x2) + (x1-x2)**2 - 1.5*x1 + 2.5*x2 + 1
surf = ax.plot_surface(x1,x2,z, rstride=1, cstride=1, cmap=cm.jet)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel("f(x1,x2)")
plt.title("McCormick function")

plt.show()
```

```python
#Matyas function
from math import pi
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')

x1 = np.linspace(-5,5,300)
x2 = np.linspace(-5,5,300)
x1, x2 = np.meshgrid(x1, x2)
z = 0.26*(x1**2 + x2**2) - 0.48*x1*x2
surf = ax.plot_surface(x1,x2,z, rstride=1, cstride=1, cmap=cm.jet)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel("f(x1,x2)")
plt.title("Matyas function")

plt.show()
```

```python
#The main difference that I could spot was in the shape and the number of local minimas that a particular test function was
#offering.

#1. The Rastrigin function has a global minima and many local minimas which were regulary distributed.

#2. Ackley function is similar to Rastrigin having many local minimas with a big hole in the centre

#3. Sphere function is a bowl shaped function that offers only a global minimum at the centre with no local minimas 

#4. Rosenbrock function, a valley shaped function also has only a global minimum lieing in its narrow, parabolic valley 
    # and no local minimas

#5. Mccormick Function is a plate shaped function with no local minimas except the global one.

#6. Matyas function is also a plate shaped function with no local minimas except the global one
```
