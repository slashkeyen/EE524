```python
#1 Plot of Rastrigin functions

import numpy as np

import matplotlib.pyplot as plt

n1 = 1
n2 = 2
k1 = 100
k2 = 100
x1 = np.linspace(-5,5,k1)
x2 = np.linspace(-5,5,k2)

f1 = (10*n1) + ((x1**2) - (10*np.cos(2*np.pi*x1)))


xx1,xx2 = np.meshgrid(x1,x2)

f2 = (10*n2) + ((xx1**2) - (10*np.cos(2*np.pi*xx1))) + ((xx2**2) - (10*np.cos(2*np.pi*xx2)))

fig = plt.figure()
ax1 = fig.add_subplot(211)
ax1.plot(x1,f1)
ax1.set_xlabel('x1 values')
ax1.set_ylabel('f(x1)')
ax1.set_title('Plot of Rastrigin function in 1-D')
ax1.grid()


ax2 = fig.add_subplot(212,projection='3d')
ax2.plot_surface(xx1,xx2,f2)
ax2.set_xlabel('x1 values')
ax2.set_ylabel('x2 values')
ax2.set_zlabel('f(x1,x2)')
ax2.set_title('Plot of Rastrigin function in 2-D')
ax2.grid()
```

```python
#3 Plot of Sphere functions

import numpy as np

import matplotlib.pyplot as plt


n1 = 1
n2 = 2
k1 = 100
k2 = 100
x1 = np.linspace(-5,5,k1)
x2 = np.linspace(-5,5,k2)

xx1,xx2 = np.meshgrid(x1,x2)
f2 = xx1**2 + xx2*2

fig = plt.figure()
ax2 = fig.add_subplot(111,projection='3d')
ax2.plot_surface(xx1,xx2,f2)
ax2.set_xlabel('x1 values')
ax2.set_ylabel('x2 values')
ax2.set_zlabel('f(x1,x2)')
ax2.set_title('Plot of Sphere function')
ax2.grid()
```

```python
#4 Plot of Rosenbrock functions

import numpy as np

import matplotlib.pyplot as plt


n1 = 1
n2 = 2
k1 = 100
k2 = 100
x1 = np.linspace(-5,5,k1)
x2 = np.linspace(-5,5,k2)

xx1,xx2 = np.meshgrid(x1,x2)
f2 = (100*(((xx1**2)-xx2)**2)) + ((1-xx1)**2)

fig = plt.figure()
ax2 = fig.add_subplot(111,projection='3d')
ax2.plot_surface(xx1,xx2,f2)
ax2.set_xlabel('x1 values')
ax2.set_ylabel('x2 values')
ax2.set_zlabel('f(x1,x2)')
ax2.set_title('Plot of Rosenbrock function')
ax2.grid()

```

```python
#5 Plot of McCormick function

import numpy as np

import matplotlib.pyplot as plt

n1 = 1
n2 = 2
k1 = 100
k2 = 100
x = np.linspace(-5,5,k1)
y = np.linspace(-5,5,k2)

xx,yy = np.meshgrid(x,x)
f2 = np.sin(xx+yy) + ((xx-yy)**2) - (1.5*xx) + (2.5*yy) + 1

fig = plt.figure()
ax2 = fig.add_subplot(111,projection='3d')
ax2.plot_surface(xx,yy,f2)
ax2.set_xlabel('x values')
ax2.set_ylabel('y values')
ax2.set_zlabel('f(x,y)')
ax2.set_title('Plot of McCormick function')
ax2.grid()
```

```python
#6 Plot of Matyas function

import numpy as np

import matplotlib.pyplot as plt

n1 = 1
n2 = 2
k1 = 100
k2 = 100
x = np.linspace(-5,5,k1)
y = np.linspace(-5,5,k2)

xx,yy = np.meshgrid(x,x)
f2 = (0.26*((xx**2)+(yy**2))) - (0.48*xx*yy)

fig = plt.figure()
ax2 = fig.add_subplot(111,projection='3d')
ax2.plot_surface(xx,yy,f2)
ax2.set_xlabel('x values')
ax2.set_ylabel('y values')
ax2.set_zlabel('f(x,y)')
ax2.set_title('Plot of Matyas function')
ax2.grid()
```
