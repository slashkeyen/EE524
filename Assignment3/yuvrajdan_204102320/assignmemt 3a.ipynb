```python
# rastrigin_function 

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 100)     
Y = np.linspace(-5, 5, 100)      

y= (X**2 - 10 * np.cos(2 * np.pi * X))  + 10
 
fig = plt.figure() 
plt.plot(X,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('rastrigin function plot in 2D')
plt.show()

X, Y = np.meshgrid(X, Y) 

Z = (X**2 - 10 * np.cos(2 * np.pi * X)) + (Y**2 - 10 * np.cos(2 * np.pi * Y)) + 20
 
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.set_title('Rastrigin finction plot in 3D')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap=cm.nipy_spectral, linewidth=0.1,antialiased=True)    

plt.show()
```

```python
# sphere function n=2

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X1 = np.linspace(-5, 5, 100)     
X2 = np.linspace(-5, 5, 100)     
X1, X2 = np.meshgrid(X1,X2) 

Z = X1**2 + X2**2
 
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X1, X2, Z, rstride=1, cstride=1, cmap=cm.nipy_spectral, linewidth=0.08,
  antialiased=True)    
ax.set_title('Sphere function plot ')
plt.show()
```

```python
# Rosenbrock function

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X1 = np.linspace(-5, 5, 100)     
X2 = np.linspace(-5, 5, 100)     
X1,X2 = np.meshgrid(X1,X2) 

Z = 100*(X1**2-X2)**2 +(1-X1)**2
 
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X1, X2, Z, rstride=1, cstride=1, cmap=cm.nipy_spectral, linewidth=0.08, antialiased=True)    
ax.set_title('Rosenbrock function plot ')
plt.show()
```

```python
# McCormick function

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 100)     
Y = np.linspace(-5, 5, 100)     
X, Y = np.meshgrid(X, Y) 

Z = np.sin(X+Y)+(X-Y)**2 -1.5*X +2.5*Y +1
 
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.nipy_spectral, linewidth=0.08,antialiased=True)    
ax.set_title('McCormick function plot ')
plt.show()
```

```python
# Matyas function

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5, 5, 100)     
Y = np.linspace(-5, 5, 100)     
X, Y = np.meshgrid(X, Y) 

Z = 0.26*(X**2 + Y**2) - 0.48*X*Y
 
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.nipy_spectral, linewidth=0.08, antialiased=True)    
ax.set_title('Matyas function plot in 3D')
plt.show()
```

All the function are continuous and differentiable . Rosenbrock is non-convex and others are convex.
