```
from mpl_toolkits import mplot3d
```

```
x1=np.arange(-10,10,.1)
```

```
x2=np.arange(-10,10,.1)
```

```
y=np.array([np.sin(x1+x2)+(x1-x2)**2-1.5*x1+2.5*x2+1])
```

```
x1,x2=np.meshgrid(x1,x2)
ax=plt.axes(projection='3d')
ax.plot_surface(x1,x2,y)
plt.show()
```
