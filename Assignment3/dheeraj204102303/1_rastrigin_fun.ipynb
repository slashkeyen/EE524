```
import matplotlib.pyplot as plt
import random
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.cm as cm
```

```
x1=np.arange(-10,10,.1)
```

```
x2=np.arange(-10,10,.1)
```

```

n=1
y1=10*n+x1**2-10*np.cos(2*np.pi*x1)
plt.plot(x1,y1)
plt.show()
```

```
n=2
y2=np.array([10*n+x1**2-10*np.cos(2*np.pi*x1)+x2**2-10*np.cos(2*np.pi*x2)])

```

```
x1,x2=np.meshgrid(x1,x2)
ax=plt.axes(projection='3d')
ax.plot_surface(x1,x2,y2)
plt.show()
```
