```python
# Importing Libraries
from matplotlib import cm  # color map
from mpl_toolkits.mplot3d import Axes3D #3d plot
import math
import matplotlib.pyplot as plt #matplotlib
import numpy as np
```

## Question no. 1

## 2D plot of Rastrigin function

```python
x =np.linspace(-5.12,5.12,200)

y= (10 + (x**2 -10*np.cos(2*3.14*x)) )

plt.plot(x,y , 'r+')
plt.title("2d plot of Rastrigin function")

plt.show()
```

## 3D plot of Rastrigin function


```python
# Rastrigin function 3d plot

X = np.linspace(-5.12, 5.12, 200)    
Y = np.linspace(-5.12, 5.12, 200)    
X, Y = np.meshgrid(X, Y)

Z = (X**2 - 10 * np.cos(2 * 3.14 * X)) +(Y**2 - 10 * np.cos(2 * 3.14 * Y)) + 20

fig = plt.figure(figsize=(10, 5))
ax = fig.gca(projection='3d') # matplotlib imported together with numpy to do some plots to returns the axes associated with the fig 

surf = ax.plot_surface(X, Y, Z,cmap=cm.hsv)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.title("3d plot of Rastrigin")
plt.show()
```

## Question no. 2

## Ackley's function for n=1

```python
x = np.arange(-32,32,0.25)

a = 20
b = 0.2

c = (-a * np.exp(-b * np.sqrt(x**2 )))
d= (-np.exp(np.cos(2*np.pi*x) ))
                   
y = a + np.exp(1) + c + d
                   
plt.plot(x,y,'k+')
plt.title("Ackley 2d plot")
plt.show()  
                       
    
```

## Ackley's funtion for n=2

```python
X = np.arange(-5, 5, 0.01)
Y = np.arange(-5,5,0.01)
X, Y = np.meshgrid(X, Y)

a = 20
b = 0.2

c = -a * np.exp(-b * np.sqrt(X*X + Y*Y) / 2) 
d = -np.exp((np.cos(2*np.pi*X) + np.cos(2*np.pi*Y)) / 2)
Z = a + np.exp(1) + c + d

fig = plt.figure(figsize=(12, 7))
ax = fig.gca(projection='3d')  # 3d plot 

surf = ax.plot_surface(X, Y, Z, cmap=cm.seismic)
fig.colorbar(surf, shrink=0.5, aspect=5)    # Add a color bar which maps values to colors

plt.title("3d plot of ackley ")
plt.show()
```

## Question no. 3

## Sphere plot

```python
# Sphere function for n=2
x = np.linspace(-5.12,5.12, num = 100)
y = np.linspace(-5.12,5.12, num = 100)
x,y = np.meshgrid(x,y)

z = x**2 + y**2

z = np.array(z)

fig = plt.figure(figsize=(12, 7))
ax = fig.gca(projection='3d')      # 3d plot 

surf = ax.plot_surface(x, y, z, cmap=cm.autumn)
fig.colorbar(surf, shrink=0.5, aspect=10)

plt.title("Sphere plot")
plt.show()
```

## Question no. 4

## Rosenbrock plot

```python
# plotting the Rosenbrock function

# Evaluate function
x = np.arange(-2, 2, 0.15)
y = np.arange(-1, 3, 0.15)
x,y = np.meshgrid(x,y)

z = (x-1)**2 + 100*(x**2-y)**2

# Plot the surface
fig = plt.figure(figsize=(12, 7))
ax = fig.gca(projection='3d')

surf = ax.plot_surface(x,y,z, cmap=cm.hot)
                      
fig.colorbar(surf, shrink=0.5, aspect=10)

plt.title("Rosenbrock plot")
plt.show()
```

## Question no. 5

## McCormick plot

```python
# plotting McCormick function
x = np.linspace(-1.5,4,100)
y = np.linspace(-3,3,100)
x,y = np.meshgrid(x,y)

a = np.sin(x+y)
b = -1.5*x+2.5*y+1

z = a + (x-y) + b

fig = plt.figure(figsize=(12, 5))
ax = fig.gca(projection='3d')

surf = ax.plot_surface(x,y,z, cmap=cm.RdYlGn)
                      
fig.colorbar(surf, shrink=0.5, aspect=10)

plt.title("McCormick plot")
plt.show()
```

## Question no. 6

## Matyas plot

```python
# plotting Matyas function
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
x,y = np.meshgrid(x,y)

a = 0.26*(x**2+y**2)
b = -0.48*x*y

z = a+b


fig = plt.figure(figsize=(12, 7))
ax = fig.gca(projection='3d')

surf = ax.plot_surface(x,y,z, cmap=cm.magma)
                      
fig.colorbar(surf, shrink = 1 ,aspect=10)

plt.title("Matyas plot")
plt.show()
```
